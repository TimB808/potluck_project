# Preprocessing

## Dropping columns

```python
new_df = df.iloc[:50]

new_df = df.drop(columns=['id', 'minutes', 'contributor_id', 'submitted', 'tags',
       'nutrition', 'n_steps', 'description', 'n_ingredients'])
```

## Lower Case

```python
def lowercase(text):
    text = text.lower()
    return text

new_df['ingredients'] = new_df['ingredients'].apply(lowercase)

new_df.head()
```

## Remove numbers

```python
def remove_num(text):
    text = ''.join(char for char in text if not char.isdigit())
    return text

new_df['ingredients'] = new_df['ingredients'].apply(remove_num)

new_df.head()
```

## Remove spaces

```python
def remove_spaces(text):
    text = text.strip()
    return text

new_df['ingredients'] = new_df['ingredients'].apply(remove_spaces)

new_df.head()
```

## Remove Stopwords

```python
# new stopwords in newStopWords variable to be edited

stopwords = nltk.corpus.stopwords.words('english')
newStopWords = ['salt', 'water', 'olive oil', 'vegetable oil', 'pepper', 'flour']
stopwords.extend(newStopWords)

def remove_stopwords(text):
    tokens = nltk.word_tokenize(text)
    new_text = [word for word in tokens if word.lower() not in stopwords]
    return ' '.join(new_text)

new_df['ingredients'] = new_df['ingredients'].apply(remove_stopwords)
new_df.head()
```

## Create list column from ingredients string column

```python
new_df['ingredients_list'] = new_df['ingredients'].apply(lambda x: x.strip('[]').split(','))
```

## Remove punctuation

```python
def remove_punct_list(text):
    cleaned_text = []
    for word in text:
        for punctuation in string.punctuation:
            word = word.replace(punctuation, '')
        cleaned_text.append(word)
    return cleaned_text
```

```python
new_df['clean_ingredients_list'] = new_df['ingredients_list'].apply(remove_punct_list)    

new_df.head()
```
